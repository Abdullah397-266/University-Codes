## Data Visualization 


### PieChart
Displays proportions of categories as slices of a pie

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%.2f%%') # autopct specifies formatting of percentage labels displayed
plt.title("Pie Chart Example")
plt.show()
```

### Violin Plot
Shows the distribution of the data and its density

```python
import numpy as np

data = np.random.normal(0, 1, 100) # mean,std, size
plt.violinplot(data)
plt.title("Violin Plot Example")
plt.show()
```

### Heatmap

```python
import numpy as np

data = np.random.rand(5, 5)
print(data)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title("Heatmap Example")
plt.colorbar()
plt.show()
```

### Area Plot

```python
days = [1, 2, 3, 4, 5]
values = [3, 8, 1, 10, 5]

plt.fill_between(days, values, color="skyblue", alpha=0.4)
plt.plot(days, values, color="purple", alpha=0.6, linewidth=2)
plt.title('Area Plot Example')
plt.show()
```

### Stacked Bar Plot

```python
categories = ['A', 'B', 'C']
data1 = [5, 7, 8]
data2 = [3, 8, 6]

plt.bar(categories, data1, color='lightblue', label='Data1')
plt.bar(categories, data2, bottom=data1, color='pink', label='Data2')
plt.legend()
plt.title('Stacked Bar Plot Example')
plt.show()
```

### Stem Plot
Displays data points with lines extending to a baseline.

```python
x = [0, 1, 2, 3]
y = [2, 4, 6, 8]

plt.stem(x, y)
plt.title("Stem Plot Example")
plt.show()
```

### Error Bar Plot
Used to show variability in data

```python
x = [1, 2, 3, 4]
y = [2.5, 3.6, 4.1, 5.8]
errors = [0.2, 0.4, 0.3, 0.2]

plt.errorbar(x, y, yerr=errors, fmt='-o')
plt.title("Error Bar Plot Example")
plt.show()
```

### Dual-Axis Plot
Allows you to plot two datasets with different y-axis scales that share the same x-axis. It is useful in situations where two variables are closely related but have different ranges or units.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g')
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b')

plt.title("Dual-Axis Plot Example")
plt.show()
```

### Subplots

```python
fig, axs = plt.subplots(2, 2)

for i in range(2):
    for j in range(2):
        axs[i, j].plot(np.random.rand(10))

plt.suptitle("Subplots Example")
plt.show()
```

# EDA

### Built-in datasets
Seaborn has multiple builtin datasets that can be listed and loaded like this

```python
import seaborn as sns

available_datasets = sns.get_dataset_names()
print(available_datasets)
```

```python
data = sns.load_dataset('titanic')
data
```

### Basic Statistics

```python
print(f"Mean Age: {data['age'].mean()}")
print(f"Median Age: {data['age'].median()}")
print(f"Mode of Age: {data['age'].mode()[0]}")
```

```python
print(f"Standard Deviation of Age: {data['age'].std()}")
print(f"Variance of Age: {data['age'].var()}")
```

```python
print(f"Skewness of Age: {data['age'].skew()}") # asymmetry of a dataset’s distribution around its mean
print(f"Kurtosis of Age: {data['age'].kurt()}") #tailedness of the distribution, helps determine if the data has outliers
```

### Univariate Analysis

#### Numerical Data: Distribution Plot

```python
sns.histplot(data['age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.show()
```

#### Categorical Data: Frequency Plot

```python
sns.countplot(x='class', data=data)
plt.title("Passenger Class Distribution")
plt.show()
```

### Categorical Data Analysis

```python
print(data['class'].value_counts())
```

### Multivariate Analysis

#### Scatter Plot

```python
sns.scatterplot(x='age', y='fare', data=data)
plt.title("Age vs Fare")
plt.show()
```
